# ML with load_boston()

```python
# sklearn toy datasets
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Load Boston house prices dataset

```python
dataBoston = load_boston()
# print(dir(dataBoston))
# DESCR, data, feature_names, filename, target
# print(dataBoston['data'][0])
# print(dataBoston['feature_names'])
# print(dataBoston['target'])

df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df['price'] = dataBoston['target']
df
```

### Checking for Correlation

```python
corr = df.corr()
corr
# sns.heatmap(corr)
plt.imshow(corr, cmap='BuPu')
plt.colorbar()
plt.xticks(np.arange(14), list(df.columns), rotation=90)
plt.yticks(np.arange(14), list(df.columns))
plt.show()
```

### Model Machine Learning Regression

```python
model = linear_model.LinearRegression()
train = df[[ 'ZN', 'CHAS', 'RM', 'DIS', 'B']]
target = df['price']
model.fit(train, target)
# model.predict(df[[ 'ZN', 'CHAS', 'RM', 'DIS', 'B']])
model.predict(df.head(1)[['ZN', 'CHAS', 'RM', 'DIS', 'B']])
```
